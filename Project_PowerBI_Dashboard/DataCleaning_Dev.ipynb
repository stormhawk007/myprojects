```python
import regex as re

def num_convertor(data,column):
    output = []
    for i in data[column]:
        if "K" in i: 
            output.append(float(re.findall("\d*\.?\d+",i)[0])*10e2)
        elif "M" in i:
            output.append(float(re.findall("\d*\.?\d+",i)[0])*10e5)
        else:
            output.append(float(re.findall("\d*\.?\d+",i)[0]))
    return output
```

```python
df['Wage'] = num_convertor(df,'Wage')

df['Value'] = num_convertor(df,'Value')
```
